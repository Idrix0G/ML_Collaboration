```python
from sklearn.preprocessing import LabelEncoder

lenc = LabelEncoder()
data['gender']=lenc.fit_transform(data['gender'])
data['Partner']=lenc.fit_transform(data['Partner'])
data['Dependents']=lenc.fit_transform(data['Dependents'])
data['PhoneService']=lenc.fit_transform(data['PhoneService'])
data['MultipleLines']=lenc.fit_transform(data['MultipleLines'])
data['InternetService']=lenc.fit_transform(data['InternetService'])
data['OnlineSecurity']=lenc.fit_transform(data['OnlineSecurity'])
data['StreamingTV']=lenc.fit_transform(data['StreamingTV'])
data['Contract']=lenc.fit_transform(data['Contract'])
data['PaperlessBilling']=lenc.fit_transform(data['PaperlessBilling'])
data['PaymentMethod']=lenc.fit_transform(data['PaymentMethod'])
data['Churn']=lenc.fit_transform(data['Churn'])
data['OnlineBackup']=lenc.fit_transform(data['OnlineBackup'])
data['StreamingMovies']=lenc.fit_transform(data['StreamingMovies'])
data['DeviceProtection']=lenc.fit_transform(data['DeviceProtection'])
data['TechSupport']=lenc.fit_transform(data['TechSupport'])

data
```

```python
from scipy import stats

corr_df = pd.DataFrame(columns=['r', 'P-value'])

for col in data:
    if pd.api.types.is_numeric_dtype(data[col]):
        r , p = stats.pearsonr(data.Churn, data[col])
        corr_df.loc[col] = [round(r, 4), round(p, 4)]
    
corr_df
```

```python
y = data.Churn
x = data.drop(['customerID','numAdminTickets','TotalCharges','PhoneService','gender','Churn'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)

from sklearn import linear_model as lr 
train_model = lr.LogisticRegression(max_iter=7043).fit(x_train, y_train)
```

```python
lr_score = train_model.score(x_test,y_test)
print(f"Model accuracy: {lr_score}")
print(f"intercept: {train_model.intercept_}")    
print(f"slope: {train_model.coef_}") 
```

```python
from sklearn.metrics import confusion_matrix

y_prediction = train_model.predict(x_test)
y_actual = y_test
cm = confusion_matrix(y_actual, y_prediction)

# heatmap visulization of confusion matrix
f, ax = plt.subplots(figsize =(7, 7))
sns.heatmap(cm, cmap='Greens', annot = True, linewidth=1, linecolor="red", fmt =".0f", ax=ax)
plt.xlabel("prediction")
plt.ylabel("actual")
plt.show()     
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB

name =  ["Nearest Neighbor","Random Forest", "Polymomial SVM", "Decision Tree","AdaBoost", "Naive bayes"]

algorithm = [
    KNeighborsClassifier(3),
    RandomForestClassifier(max_depth=5, n_estimators=100),
    SVC(kernel="linear", C=0.025),
    DecisionTreeClassifier(max_depth=5),
    AdaBoostClassifier(n_estimators=100),
    GaussianNB()]
```

```python
#df = pd.DataFrame(columns=['Test Algorithm','Model Score'], dtype=object)
scores = []
for algorithm in (algorithm):
    algorithm.fit(x_train, y_train)
    score = algorithm.score(x_test, y_test)
    scores.append(score)
    #df.loc[col] = [name,scores]
scores
```

```python
df = pd.DataFrame()
df['Classifier'] = name
df['Accuracy Score'] = scores
df
```
